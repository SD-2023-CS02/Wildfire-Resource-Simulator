```python
import pandas as pd

# read output gdb table from gdb to csv converter
# TTR (Time to Read) is about 35< seconds for 2020 data, about 50< for 2021
gdb_operational_data_2020 = pd.read_csv('output/output_op_data_2020.csv', low_memory=False)
gdb_operational_data_2021 = pd.read_csv('output/output_op_data_2021.csv', low_memory=False)
```

```python
# display types from table
print('2020 DataTypes:', gdb_operational_data_2020.dtypes.size)
print('2021 Datatypes:', gdb_operational_data_2021.dtypes.size)

print(gdb_operational_data_2021.dtypes)
```

```python
# copy data as Pandas DataFrame, clean it
cleaned_op_data_2020 = pd.DataFrame(gdb_operational_data_2020)
cleaned_op_data_2021 = pd.DataFrame(gdb_operational_data_2021)

# columns to remove
to_clean_2020 = ['IRWINID', 'GeometryID', 'ArchClearance', 'Unnamed: 0', 'RepairComments', 'DeleteThis', 'LocalIncidentID', 'GlobalID', 'GACC', 'LineDateTime', 'ComplexName', 'geometry']
to_clean_2021 = ['Unnamed: 0', 'DeleteThis', 'FeatureAccess', 'Label', 'GlobalID', 'LineWidthFeet', 'LineDateTime', 'Shape_Length', 'ArchClearance', 'IRWINID', 'LengthFeet', 'IsVisible']
cleaned_op_data_2021.drop(to_clean_2021, axis=1, inplace=True)
cleaned_op_data_2020.drop(to_clean_2020, axis=1, inplace=True)

flight_data_2020 = cleaned_op_data_2020[cleaned_op_data_2020['Comments'].str.contains('Flight', na=False)]
flight_data_2021 = cleaned_op_data_2021[cleaned_op_data_2021['FeatureCategory'].str.contains('Flight', na=False)]

display(flight_data_2020)
# display(cleaned_op_data_2021)
# print(cleaned_op_data.dtypes)
```
