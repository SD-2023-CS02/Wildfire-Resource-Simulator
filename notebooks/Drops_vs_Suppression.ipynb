```python
import pandas as pd


stats_df = pd.read_csv('../output/suppression_stats.csv').sort_values('FireDiscoveryDateTime')
stats_df = stats_df.loc[stats_df['FireDiscoveryDateTime'] >= '2020-01-01'].loc[stats_df['FireDiscoveryDateTime'] < '2023-01-01']
drops_df = pd.read_csv('../output/plane_drops_results.csv')

display(stats_df)
display(drops_df)
```

```python
fires_with_drops = set()

for i in range(len(drops_df)):
    row = drops_df.iloc[i]
    fires_with_drops.add(row['fire_id'])

print(len(fires_with_drops))
```

```python
has_drop = [stats_df.iloc[i]['UniqueFireIdentifier'] in fires_with_drops for i in range(len(stats_df))]
stats_df.loc[:, 'HasPlaneDrop'] = has_drop

stats_df.to_csv('../output/suppression_stats_with_drops.csv', index=False)

values = stats_df.value_counts(['HasPlaneDrop'])
print(values)
```
